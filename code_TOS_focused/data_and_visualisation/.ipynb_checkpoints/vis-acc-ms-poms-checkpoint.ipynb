```python
pom_uv_path = 'data_acc_uv//'

pom00 = vis.import_UV(pom_uv_path+'DJK3-015a_UV_0eq.csv', "")
pom05 = vis.import_UV(pom_uv_path+'DJK3-015b_UV_5eq.csv', "")
pom07 = vis.import_UV(pom_uv_path+'DJK3-015c_UV_7eq.csv', "")
pom09 = vis.import_UV(pom_uv_path+'DJK3-015d_UV_9eq.csv', "")
pom11 = vis.import_UV(pom_uv_path+'DJK3-015e_UV_11eq.csv', "")
pom13 = vis.import_UV(pom_uv_path+'DJK3-015f_UV_13eq.csv', "")
```
