```python
import spectra_visualiser as vis
import matplotlib.pyplot as plt
import matplotlib
from copy import deepcopy
```

# Comparison of Pure Vaterite and Aragonite Samples

```python
path = 'data_acc_ir\\'

s061a = vis.import_IR(path+'DJK2-061a_IR.csv', "DJK2-061a")
s061b = vis.import_IR(path+'DJK2-061b_IR.csv', "DJK2-061b")
s062a = vis.import_IR(path+'DJK2-062a_IR.csv', "DJK2-062a")
s062b = vis.import_IR(path+'DJK2-062b_IR.csv', "DJK2-062b")
s066a = vis.import_IR(path+'DJK2-066a_IR.csv', "DJK2-066a")
s067a = vis.import_IR(path+'DJK2-067a_IR.csv', "DJK2-067a")
s068a = vis.import_IR(path+'DJK2-068a_IR.csv', "DJK2-068a")

s061a2 = deepcopy(s061a)
s061b2 = deepcopy(s061b)
s062a2 = deepcopy(s062a)
s062b2 = deepcopy(s062b)
s066a2 = deepcopy(s066a)
s067a2 = deepcopy(s067a)
s068a2 = deepcopy(s068a)
```

```python
fig,ax = plt.subplots(2,1, figsize=(10,6), sharex=True)

ax[0].set_xlim(1200,600)
ax[0].set_ylim(20, 105)
ax[1].set_xlim(1200,600)
ax[1].set_ylim(10, 95)

ax[1].set_xlabel('Wavenumbers (cm$^{-1}$)', fontsize=14)

ax[0].plot(s061b2.data['Wavenumbers(cm-1)'], s061b2.data['Transmittance'], color='k')
ax[1].plot(s066a2.data['Wavenumbers(cm-1)'], s066a2.data['Transmittance'], color='k')

print(ax[1].get_ylim())

for n in [0]:
    y1,y2 = ax[n].get_ylim()
    ax[n].plot([1088,1088], [y1,95.65103], color='r', linestyle=':')
    ax[n].plot([876,876], [y1,44.97179], color='r', linestyle=':')
    ax[n].plot([849,849], [y1,92.91692], color='r', linestyle=':')
    ax[n].plot([744,744], [y1,88.84203], color='r', linestyle=':')
    
for n in [1]:
    y1,y2 = ax[1].get_ylim()
    ax[1].plot([1083,1083], [y1,62.26984], color='b', linestyle=':')
    ax[1].plot([852,852], [y1,22.01612], color='b', linestyle=':')
    ax[1].plot([844,844], [y1,35.32293], color='b', linestyle=':')
    ax[1].plot([713,713], [y1,42.31789], color='b', linestyle=':')
    ax[1].plot([700,700], [y1,45.35384], color='b', linestyle=':')
    
fig.supylabel('Transmittance', x=0.07, fontsize=14)

ax[0].annotate('Vaterite', fontsize=16,
            xy=(610,27), xycoords='data',
            xytext=(0,0), textcoords='offset points',
            arrowprops=dict(facecolor='black', shrink=0.05),
            bbox={'facecolor':'red', 'alpha':0.15, 'edgecolor':'red'},
            horizontalalignment='right', verticalalignment='bottom')

ax[1].annotate('Aragonite', fontsize=16,
            xy=(610,17), xycoords='data',
            xytext=(0,0), textcoords='offset points',
            arrowprops=dict(facecolor='black', shrink=0.05),
            bbox={'facecolor':'blue', 'alpha':0.15, 'edgecolor':'blue'},
            horizontalalignment='right', verticalalignment='bottom')

fig.savefig('vis-output\\ca-polymorph-ir.png')
```
