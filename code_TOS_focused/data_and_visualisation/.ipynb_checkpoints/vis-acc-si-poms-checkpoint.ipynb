```python
import spectra_visualiser as vis
import matplotlib.pyplot as plt
import matplotlib
from copy import deepcopy
```

# Comparison of POM UV Spectra

```python
path = 'data_acc_uv\\'

eq00 = vis.import_UV(path+'DJK3-015a_UV_0eq.csv', "")
eq05 = vis.import_UV(path+'DJK3-015b_UV_5eq.csv', "")
eq07 = vis.import_UV(path+'DJK3-015c_UV_7eq.csv', "")
eq09 = vis.import_UV(path+'DJK3-015d_UV_9eq.csv', "")
eq11 = vis.import_UV(path+'DJK3-015e_UV_11eq.csv', "")
eq13 = vis.import_UV(path+'DJK3-015f_UV_13eq.csv', "")
```

```python
fig,ax = plt.subplots(2,1, figsize=(10,9), sharex=True)

ax[0].set_xlim(300,850)
#ax[0].set_ylim(20, 105)
ax[1].set_xlim(300,850)
#ax[1].set_ylim(10, 95)

ax[1].set_xlabel('Wavelength (nm)', fontsize=14)

ax[0].plot(eq13.data['Wavelength(nm)'], eq13.data['Absorbance(au)'], color='#5EC0F0', label='13 eq. HCl')
ax[0].plot(eq11.data['Wavelength(nm)'], eq11.data['Absorbance(au)'], color='#648fff', label='11 eq. HCl')
ax[0].plot(eq09.data['Wavelength(nm)'], eq09.data['Absorbance(au)'], color='#785ef0', label='9 eq. HCl')
ax[0].plot(eq07.data['Wavelength(nm)'], eq07.data['Absorbance(au)'], color='#dc267f', label='7 eq. HCl')
ax[0].plot(eq05.data['Wavelength(nm)'], eq05.data['Absorbance(au)'], color='#fe6100', label='5 eq. HCl')
ax[0].plot(eq00.data['Wavelength(nm)'], eq00.data['Absorbance(au)'], color='#ffb000', label='0 eq. HCl')

ax[1].plot(eq13.data['Wavelength(nm)'], eq13.data['Relative Absorbance(au)'], color='#5EC0F0', label='13 eq. HCl')
ax[1].plot(eq11.data['Wavelength(nm)'], eq11.data['Relative Absorbance(au)'], color='#648fff', label='11 eq. HCl')
ax[1].plot(eq09.data['Wavelength(nm)'], eq09.data['Relative Absorbance(au)'], color='#785ef0', label='9 eq. HCl')
ax[1].plot(eq07.data['Wavelength(nm)'], eq07.data['Relative Absorbance(au)'], color='#dc267f', label='7 eq. HCl')
ax[1].plot(eq05.data['Wavelength(nm)'], eq05.data['Relative Absorbance(au)'], color='#fe6100', label='5 eq. HCl')
ax[1].plot(eq00.data['Wavelength(nm)'], eq00.data['Relative Absorbance(au)'],  color='#ffb000', label='0 eq. HCl')

y01,y02 = ax[0].get_ylim()
y11,y12 = ax[1].get_ylim()

ax[0].plot([440,440], [y01,y02], color='k', linestyle=':')
ax[0].plot([761,761], [y01,y02], color='k', linestyle=':')
ax[1].plot([440,440], [y11,0.8*y12], color='k', linestyle=':')
ax[1].plot([761,761], [y11,0.8*y12], color='k', linestyle=':')
       
ax[0].set_ylabel('Absorbance (au)', x=0.07, fontsize=14)
ax[1].set_ylabel('Relative Absorbance', x=0.07, fontsize=14)

ax[0].legend(loc='upper left', fontsize=12)

ax[1].annotate('{Mo$_{132}$}', fontsize=16,
            xy=(440,0.85*y12), xycoords='data',
            xytext=(0,0), textcoords='offset points',
            arrowprops=dict(facecolor='black', shrink=0.05),
            bbox={'facecolor':'brown', 'alpha':0.15, 'edgecolor':'black'},
            horizontalalignment='center', verticalalignment='bottom')

ax[1].annotate('{Mo$_{154}$}', fontsize=16,
            xy=(761,0.85*y12), xycoords='data',
            xytext=(0,0), textcoords='offset points',
            arrowprops=dict(facecolor='black', shrink=0.05),
            bbox={'facecolor':'blue', 'alpha':0.15, 'edgecolor':'black'},
            horizontalalignment='center', verticalalignment='bottom')

fig.tight_layout()

fig.savefig('vis-output\\poms-comparison.png')
```
