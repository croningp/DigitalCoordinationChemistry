## [Co(Ac2O)2OAc]X2 1H NMR

```python
fig,ax = plt.subplots(1,1, figsize=(8.5,5.5))
ax.set_xlim(6,0)
ax.set_ylim(-5,105)

color = 'k'

ax.plot(cce1_27.data['Shift(ppm)'], cce1_27.data['Relative Intensity'], linewidth=1.5, color=color)

inset = fig.add_axes([0.55,0.55,0.40,0.40])
inset.set_xlim(2.55,2.15)
inset.set_ylim(-1, 16)
inset.plot(cce1_27.data['Shift(ppm)'], cce1_27.data['Relative Intensity'], color=color, linewidth=1.5)

ax.tick_params(axis='both', labelsize=14)
inset.tick_params(axis='both', labelsize=14)

vis.core.shade_regions(ax, [[2.55,2.15,'k',0.1]])
vis.core.shade_regions(inset, [[2.55,2.15,'k',0.1]])

ax.set_xlabel('$\delta$ (ppm)', fontsize=16, fontweight='bold')
ax.set_ylabel('Relative Intensity', fontsize=16, fontweight='bold')

fig.tight_layout()

fig.savefig('output_vis\\vis-cbx-si-anhydrides-HeterolepticComplex.png')
```
