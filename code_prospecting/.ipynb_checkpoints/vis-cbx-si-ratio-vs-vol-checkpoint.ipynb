# vis-cbx-si-ratio-vs-vol

```python
import packages.carboxy.carboxy as cbx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

## Generate Random Uniform Samples in Ratio Domain

```python
# define exploration object
cce = cbx.explore.CCE(
    expl_name='', package='random',
    prev_res=None,
    kappa=6, double_weighting=True
)

# algorithmically generate cluster and template ratios, and total volume
df_ratios = cce.generate_new_values()

# convert ratios to volumes
df_volumes = cce.calculate_volumes()

# transcribe volumes to ClusterBot-legible .csv format 
df_legible = cce.write_legible_df()
```

```python
# fill array
arr = df_volumes.to_numpy()
arr_all = arr
```

```python
print(arr_all.shape)

for _ in range(0,208):
    
    # define exploration object
    cce = cbx.explore.CCE(
        expl_name='', package='random',
        prev_res=None,
        kappa=6, double_weighting=True
    )

    # algorithmically generate cluster and template ratios, and total volume
    df_ratios = cce.generate_new_values()

    # convert ratios to volumes
    df_volumes = cce.calculate_volumes()

    # transcribe volumes to ClusterBot-legible .csv format 
    df_legible = cce.write_legible_df()
    
    arr = df_volumes.to_numpy()
    arr_all = np.vstack((arr_all,arr))
    
    print(arr_all.shape)
```

```python
df = pd.DataFrame(arr_all)
df.columns = df_volumes.columns
df
```

```python
cbx.matrix.CCE_matrix(
    df, range_cluster=5.3, range_template=4.8, range_solv=6.0,
    prev_res=None
)
```

```python
print(arr_all.shape)

for _ in range(0,208):
    
    # define exploration object
    cce = cbx.explore.CCE(
        expl_name='', package='random',
        prev_res=None,
        kappa=6, double_weighting=True
    )

    # algorithmically generate cluster and template ratios, and total volume
    df_ratios = cce.generate_new_values()

    # convert ratios to volumes
    df_volumes = cce.calculate_volumes()

    # transcribe volumes to ClusterBot-legible .csv format 
    df_legible = cce.write_legible_df()
    
    arr = df_volumes.to_numpy()
    arr_all = np.vstack((arr_all,arr))
    
    print(arr_all.shape)
```
