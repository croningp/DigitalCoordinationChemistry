### NMR: {Co3O} and {Co3O(OH)} in CD2Cl2

```python
# setup figures
fig, ax = plt.subplots(2, 1, figsize=(10,4), sharex='col')
fig.subplots_adjust(hspace=0.1, wspace=0.1)

# FORMAT: spines
for i in ['top','bottom','right','left']:
    ax[0].spines[i].set_visible(False)
for i in ['top','right','left']:
    ax[1].spines[i].set_visible(False)

# FORMAT: ticks
ax[0].tick_params(axis="y", which="both", left=False, right=False, labelleft=False)
ax[0].tick_params(axis="x", which="both", bottom=False)
ax[1].tick_params(axis="y", which="both", left=False, right=False, labelleft=False)

# x and y limits
ax[0].set_xlim(9.50,7.50)
ax[0].set_ylim(-1,10)
ax[1].set_ylim(-1,17)

# plot data
ax[0].plot(co3o.data['Shift(ppm)'], co3o.data['Relative Intensity'], linewidth=1.2, color='k')
ax[1].plot(co3ooh.data['Shift(ppm)'], co3ooh.data['Relative Intensity'], linewidth=1.2, color='k')

# annotations
ax[1].set_xlabel("Chemical Shift (ppm)")
ax[0].annotate('{Co$_{3}$O} CD$_{2}$Cl$_{2}$', xy=(9.50,10), fontsize=18, fontweight='normal', color='k')
ax[1].annotate('{Co$_{3}$O(OH)} CD$_{2}$Cl$_{2}$', xy=(9.50,17), fontsize=18, fontweight='normal', color='k')

ax[0].annotate('a', xy=(8.95,8.5), fontsize=16, fontweight='bold', color='#dc267f')
ax[0].annotate('c', xy=(8.12,5), fontsize=16, fontweight='bold', color='#dc267f')
ax[0].annotate('b', xy=(7.74,7), fontsize=16, fontweight='bold', color='#dc267f')

ax[1].annotate('a', xy=(9.24,8.5), fontsize=16, fontweight='bold', color='#fe6100')
ax[1].annotate('a', xy=(8.78,16.5), fontsize=16, fontweight='bold', color='#648fff')
ax[1].annotate('c', xy=(8.29,5), fontsize=16, fontweight='bold', color='#fe6100')
ax[1].annotate('c', xy=(8.16,9), fontsize=16, fontweight='bold', color='#648fff')
ax[1].annotate('b', xy=(7.93,7), fontsize=16, fontweight='bold', color='#fe6100')
ax[1].annotate('b', xy=(7.77,13), fontsize=16, fontweight='bold', color='#648fff')

plt.tight_layout()
fig.savefig('output_vis\\vis-cbx-si-degradation-TrinuclearClustersComparison.png')
```
