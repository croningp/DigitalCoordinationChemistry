# ISE_2 Exploration

```python
# import necessary libraries
import packages.carboxy.carboxy as cbx
```

```python
# define exploration object
path = 'data_synth\\'

ise2 = cbx.explore.ISE(
    expl_name='ISE_2', package='skopt',
    prev_res=path+'ISE_1_result_AMF.csv',
    kappa=6, double_weighting=True
)
print('Object created\n')

# algorithmically generate cluster and template ratios, and total volume
df_ratios = ise2.generate_new_values()
print('New samples generated\n')

# convert ratios to volumes
df_volumes = ise2.calculate_volumes()
print('Conversion to volumes complete\n')

# transcribe volumes to ClusterBot-legible .csv format 
df_legible = ise2.write_legible_df()
print('Conversion to ClusterBot legible format complete')
```

```python
cbx.matrix.ISE_matrix(
    df_volumes, range_cluster=6, range_template=4.8, range_solv=6.0,
    prev_res=path+'ISE_1_legible.csv'
)
```

```python
cbx.matrix.ISE_matrix_concise(
    df_ratios,
    prev_res_ratio=path+'ISE_1_ratio.csv'
)
```
