## m/z Similarity

```python
def mz_similarity(spectrum1, spectrum2, noise_threshold):
    df = cbx.msrep.gen_table(spectrum1,spectrum2)
    
    actual_peak = []
    only_1 = []
    only_2 = []

    for index, row in df.iterrows():
        if (row['Relative Intensity 1'] >= noise_threshold) or (row['Relative Intensity 2'] >= noise_threshold):
            actual_peak.append(row['m/z'])
        if (row['Relative Intensity 1'] >= noise_threshold) and (row['Relative Intensity 2'] == 0.0):
            only_1.append(row['m/z'])
        if (row['Relative Intensity 1'] == 0.0) and (row['Relative Intensity 2'] >= noise_threshold):
            only_2.append(row['m/z'])

    score = round((len(actual_peak)-len(only_1)-len(only_2))/len(actual_peak) * 100, 2)
    return score
```
