# vis-cbx-si-M10

```python
import packages.spectra_visualiser.spectra_visualiser as vis
import matplotlib.pyplot as plt
```

```python
ise = vis.import_IR("data_other\\DJK3-001a_IR_M10 from ISE.csv", "")
sth = vis.import_IR("data_other\\DJK3-001b_IR_Fe10.csv", "")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,6))

ax.set_xlim(4050,450)
ax.set_ylim(20,105)

vis.core.feature_scale_y(ise, y_label='Transmittance')
vis.core.feature_scale_y(sth, y_label='Transmittance')

line1, = ax.plot(sth.data['Wavenumbers(cm-1)'], sth.data['Transmittance'], color='#dc267f', linewidth=1.5)
line1.set_label('{Fe$_{10}$}')
line2, = ax.plot(ise.data['Wavenumbers(cm-1)'], ise.data['Transmittance'], color='#648fff', linewidth=1.5)
line2.set_label('{M$_{10}$} from ISE')

inset = fig.add_axes([0.2,0.2,0.4,0.4])
inset.set_xlim(710,500)
inset.set_ylim(10, 95)
inset.plot(sth.data['Wavenumbers(cm-1)'], sth.data['Relative Transmittance'], color='#dc267f', linewidth=1.5)
inset.plot(ise.data['Wavenumbers(cm-1)'], ise.data['Relative Transmittance'], color='#648fff', linewidth=1.5)
ax.set_xlabel('Wavenumbers', fontsize=16, fontweight='bold')
ax.set_ylabel('Transmittance', fontsize=16, fontweight='bold')

ax.tick_params(axis='both', labelsize=14)
inset.tick_params(axis='both', labelsize=14)

vis.core.shade_regions(ax, [[500,710,'k',0.1]])
vis.core.shade_regions(inset, [[500,710,'k',0.1]])

ax.legend(loc='lower right', fontsize=16)

fig.savefig('output_vis//vis-cbx-si-M10.png')
```
