# CCE_4 || Exploration

```python
# import necessary libraries
import packages.carboxy.carboxy as cbx
```

```python
# define exploration object
path = 'data_synth\\'

cce4 = cbx.explore.CCE(
    expl_name='CCE_4', package='skopt',
    prev_res=path+'CCE_3_result_AMF_EDIT.csv',
    kappa=6, double_weighting=True
)
print('Object created\n')

# algorithmically generate cluster and template ratios, and total volume
df_ratios = cce4.generate_new_values()
print('New samples generated\n')

# convert ratios to volumes
df_volumes = cce4.calculate_volumes()
print('Conversion to volumes complete\n')

# transcribe volumes to ClusterBot-legible .csv format 
df_legible = cce4.write_legible_df()
print('Conversion to ClusterBot legible format complete')
```

```python
cbx.matrix.CCE_matrix(
    df_volumes, range_cluster=6.2, range_template=6.1, range_solv=6.1,
    prev_res=path+'CCE_3_legible.csv'
)
```

```python
cbx.matrix.CCE_matrix_concise(
    df_ratios,
    prev_res=path+'CCE_2_ratio.csv',
    drop_metadata=False
)
```
